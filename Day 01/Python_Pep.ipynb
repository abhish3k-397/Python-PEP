```python
a = int(input("Enter your first number: "))
b = int(input("Enter your second number: "))
op = input("Enter the operation you want to do (+,-,*,/,//): ")
if op == "+":
  print(a+b)
elif op == "-":
  print(a-b)
elif op == "*":
  print(a*b)
elif op == "/":
  print(a/b)
elif op == "//":
  print(a//b)
```

continuous input


```python
print("Enter the String of integers and operations: ")
a = int(input())
while True:
  op = input()
  if op == "end":
    break
  b = int(input())
  if op == "+":
    a = a+b
  elif op == "-":
    a = a-b
  elif op == "*":
    a = a*b
  elif op == "/":
    a = a/b
print(a)
```

Prime checker


```python
n = int(input())
if n <2:
  print("Not prime")
else:
  prime = True

  for i in range(2,n//2+1):
    if n%i == 0:
      prime = False
      break
  if prime:
    print("Prime")
  else:
    print("Not prime")
```
