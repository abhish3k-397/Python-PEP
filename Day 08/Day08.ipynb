# ADVANCED FUNCTIONS

Lambda


```python
sum = lambda x,y:x+y
sum(10,29)
```

```python
multiply = lambda x,y:x*y
multiply(10,29)
```

```python
sub = lambda x,y:x-y
sub(10,29)
```

```python
divide = lambda x,y:x/y
divide(10,29)
```

```python
largest = lambda x,y: max(x,y)
largest(10,29)
```

```python
smallest = lambda x,y: min(x,y)
smallest(23,343)
```

```python
n = int(input("Enter to check prime: "))
prime = lambda x: n > 1 and all(n % i != 0 for i in range(2, int(n**0.5) + 1))
prime(23)
```

```python
c = int(input("Enter celsius: "))
fahrenheit = lambda x: (x * 1.8) + 32
fahrenheit(c)
```

```python
n = int(input("Enter the number to get table: "))
table = lambda n: list(map(lambda i: print(f"{n} * {i} = {n*i}"),range(1,11)))
_ = table(n)
```

```python
n = int(input("Enter number to get cube: "))
cube = lambda n: n**3
cube(n)
```

```python
n = int(input("Enter number to check even odd: "))
even_odd = lambda n: 'even' if n%2 == 0 else 'odd'
even_odd(n)
```

```python
n = int(input("Enter number to check polarity: "))
polarity = lambda n: 'positive' if n > 0 else 'zero' if n == 0 else 'negative'
polarity(n)
```

```python
prices = list(map(int,input("Enter prices: ").split()))
discount = list(map(lambda n : int(n*0.8),prices))
print(discount)
```

```python
n = int(input("Enter the number to get table: "))
table = lambda n: list(map(lambda i: print(f"{n} + {i} = {n+i}"),range(1,11)))
_ = table(n)
```

```python
n = list(map(str,input("Enter text: ").split()))
caps = lambda n: [i.upper() for i in n]
caps(n)
```

```python
lst = list(map(int,input("Enter the numbers: ").split()))
filt = list(filter(lambda n: n>6767,lst))
print(filt)
```
