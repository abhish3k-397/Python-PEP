Q1. Accept 5 string store as tuple print with indices


---

Q2. Print the same in reverse order

````python
tup = ()
for i in range(5):
  item = input(f"Enter string {i+1}: ")
  tup += (item,)
print("Tuple elements with indices:")
for index, value in enumerate(tup):
  print(f"Index {index}: {value}")
````

````python
tup = ()
for i in range(5):
  item = input(f"Enter string {i+1}: ")
  tup += (item,)
print("Tuple elements in reverse order:")
for index, value in enumerate(reversed(tup)):
  print(5 - index, value)
````

# Dictionary

| Function    | Description                          |
| ----------- | ------------------------------------ |
| `len(d)`    | Number of key-value pairs            |
| `type(d)`   | Returns the type                     |
| `str(d)`    | Converts to a string                 |
| `dict()`    | Creates a dictionary                 |
| `sorted(d)` | Returns a sorted list of keys        |
| `list(d)`   | Returns a list of keys               |
| `max(d)`    | Largest key                          |
| `min(d)`    | Smallest key                         |
| `sum(d)`    | Sum of keys (if numeric)             |
| `any(d)`    | `True` if at least one key is truthy |
| `all(d)`    | `True` if all keys are truthy        |






| Method         | Syntax                       | Purpose                                                                    |
| -------------- | ---------------------------- | -------------------------------------------------------------------------- |
| `clear()`      | `d.clear()`                  | Removes all key-value pairs                                                |
| `copy()`       | `d.copy()`                   | Returns a shallow copy of the dictionary                                   |
| `fromkeys()`   | `dict.fromkeys(keys, value)` | Creates a new dictionary with specified keys and a common value            |
| `get()`        | `d.get(key, default)`        | Returns the value for a key, or a default if the key doesn't exist         |
| `items()`      | `d.items()`                  | Returns a view of all `(key, value)` pairs                                 |
| `keys()`       | `d.keys()`                   | Returns a view of all keys                                                 |
| `pop()`        | `d.pop(key[, default])`      | Removes the specified key and returns its value                            |
| `popitem()`    | `d.popitem()`                | Removes and returns the last inserted `(key, value)` pair                  |
| `setdefault()` | `d.setdefault(key, default)` | Returns the value of the key; if absent, inserts it with the default value |
| `update()`     | `d.update(other)`            | Adds or updates key-value pairs from another mapping or iterable           |
| `values()`     | `d.values()`                 | Returns a view of all values                                               |


````python
student = {
    "Name":"Abhi",
    "Age":21,
    "CGPA":8.29
}

print(student)
print(*student,*student.values())
for i in student:
  print(i,student[i])
  print(i,student.get(i))

#ADD stuff
student["city"] = "Tirur"
print(student)
#Update Stuff

student.update({"CGPA":8.99})
print(student)
````
