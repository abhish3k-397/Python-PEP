| Style                                | Syntax                             | Example                              |
| ------------------------------------ | ---------------------------------- | ------------------------------------ |
| 1. Import entire module              | `import module`                    | `import math`                        |
| 2. Import with alias                 | `import module as alias`           | `import numpy as np`                 |
| 3. Import specific object            | `from module import name`          | `from math import sqrt`              |
| 4. Import multiple objects           | `from module import a, b`          | `from math import sin, cos, tan`     |
| 5. Import everything (wildcard)      | `from module import *`             | `from math import *`                 |
| 6. Import specific object with alias | `from module import name as alias` | `from math import factorial as fact` |
