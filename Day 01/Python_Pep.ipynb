continuous input


```python
print("Enter the String of integers and operations: ")
a = int(input())
while True:
  op = input()
  if op == "end":
    break
  b = int(input())
  if op == "+":
    a = a+b
  elif op == "-":
    a = a-b
  elif op == "*":
    a = a*b
  elif op == "/":
    a = a/b
print(a)
```

Prime checker


```python
n = int(input())
if n <2:
  print("Not prime")
else:
  prime = True

  for i in range(2,n//2+1):
    if n%i == 0:
      prime = False
      break
  if prime:
    print("Prime")
  else:
    print("Not prime")
```

Coupon Problem

```python
import random
random.seed(4545)
days_active = random.randint(0,1000)
lifetime_orders = random.randint(0,1000)
new_user = random.choice([True,False])
returning_user = random.choice([True,False])
print(days_active)
print(lifetime_orders)
print(new_user)
print(returning_user)
if new_user:
  print("New user discout 50% flat on first 5 orders *off on delivery fees")
elif days_active == 500 or days_active == 1000:
  print("Milestone reward 500RS off on order above 10000")
elif lifetime_orders == 500 or lifetime_orders == 1000:
  print("Milestone reward 50RS off on order above 1000")
elif returning_user:
  print("Welcome back. 99% OFF !!!! upto 40 on orders above 10000000")
```

ATM


```python
acc = int(input("Enter your account number: "))
account_number = 1234
p = int(input("Enter your pin: "))
pin = 123
if acc == account_number and p == pin:
  print("Welcome to ATM")
  action = input("withdraw , deposit , checkbalance : ")
  balance = 100000
  amount = int(input("Enter the amount: "))
  if action == "withdraw":
    if balance >= amount:
      balance = balance - amount
      print("Successful!, Your new balance is",balance)
    else:
      print("Insufficient balance")
  elif action == "deposit":
    balance = balance + amount
    print("Your balance is",balance)
  elif action == "checkbalance":
    print("Your balance is",balance)
  else:
    print("Invalid action")
else:
  print("Invalid Credentials")
```
