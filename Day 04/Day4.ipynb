```python
n = int(input())
re = 0
while(n>0):
  d = n%10
  re *= 10
  re += d
  n//=10
print(re)
```

```python
n = int(input())
re = 0
num = n
while(n>0):
  d = n%10
  re *= 10
  re += d
  n//=10
if(re == num):
  print("Palindrome")
else:
  print("Not Palindrome")
```

# Nested Loops


```python
for i in range(3):
  for j in range(3):
    print(f"{i} {j}")
```

```python
for i in range (1,4):
  print(f"Row {i}")
  for j in range(1,4):
    print(f"\t Column {j}")
```

```python
for i in range(3):
  for j in range(4):
    print("* ",end="")
  print()
```

```python
for i in range(3):
  for j in range(1,5):
    print(f"{j} ",end="")
  print()
```

```python
for i in range(3):
  s = "A"
  for j in range(4):
    print(f"{s} ",end="")
    s = chr(ord(s)+1)
  print()
```

```python
for i in range(1,4):
  for j in range(1,4):
    print(f"{i*j} ",end="")
  print()
```

```python
for i in range(1,5):
  for j in range(1,5):
    print(f"{i} ",end="")
  print()
```

```python
n = 1
for i in range(3):
  for j in range(3):
    print(f"{n} ",end="")
    n+=1
  print()
```

```python
n = 2
for i in range(3):
  for j in range(3):
    print(f"{n} ",end="")
    n+=2
  print()
```

```python
for i in range(3):
  for j in range(4):
    print(i,j,end="")
  print()
```
