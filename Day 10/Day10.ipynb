Q6. email validation

```python
email = input("Enter email: ")
print(email.split("@")[1])

if email.count("@") == 1 and "." in email.split("@")[1]:
    print("Valid Email")
else:
    print("Invalid Email")
```

```python
import re

email = input("Enter email: ")

pattern = r'^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$'

if re.match(pattern, email):
    print("Valid Email")
else:
    print("Invalid Email")
```
