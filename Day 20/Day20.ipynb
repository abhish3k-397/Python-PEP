```python
import numpy as np

students = np.array([
    "Rahul",
    "Priya",
    "Aman",
    "Neha",
    "Arjun",
    "Sneha",
    "Karan",
    "Anjali"
])

subjects = np.array([
    "CSE101",
    "MAT102",
    "PHY103",
    "ENG104",
    "DBM105"
])

marks = np.array([
    [85, 92, 78, 88, 91],
    [76, 81, 69, 74, 80],
    [90, 88, 94, 91, 89],
    [65, 70, 68, 75, 72],
    [88, 84, 80, 86, 90],
    [72, 79, 75, 80, 77],
    [91, 95, 89, 90, 94],
    [82, 78, 85, 81, 79]
])


print("Name".ljust(10),end="")

for subject in subjects:
  print(subject.center(8),end="")

print()
for i in range(len(students)):
  print(students[i].ljust(10),end="")
  for j in range(len(subjects)):
    print(str(marks[i][j]).center(8),end="")
  print()
```

```python
avg = np.mean(marks, axis=1)
for i in range(len(students)):
  print(students[i].ljust(10),end="")
  print(f"{avg[i]:.2f}")
```

```python
highest = np.max(marks, axis = 1)
lowest   = np.min(marks, axis = 1)

for name,high,low in zip(students,highest,lowest):
  print(f"{name.ljust(10)} Highest = {high} Lowest = {low}")

print()
```

```python
std = np.std(marks, axis=1)

for name,s in zip(students, std):
  print(f"{name.ljust(10)} Standard Deviation = {s:.2f}")
```

```python
ranks = np.argsort(avg)[::-1]

# for rank,name in zip(ranks,students):
#   print(f"Rank = {rank} {name.ljust(10)}")
i = 0
for rank,index in enumerate(ranks,start = 1):
  print(f"Rank = {rank} {students[index].ljust(10)}",avg[i])
  i += 1
```

```python
class_avg = np.mean(avg)

print("Class Average =".ljust(15), class_avg)

for i in range(len(students)):
    if avg[i] > class_avg:
        print(students[i].ljust(15), avg[i])
```

```python
subject_avg = np.mean(marks, axis=0)

for i in range(len(subjects)):
    print(subjects[i], subject_avg[i])
```

```python
subject_high = np.max(marks, axis=0)
subject_low = np.min(marks, axis=0)

for i in range(len(subjects)):
    print(f"{subjects[i]}  Highest={subject_high[i]}  Lowest={subject_low[i]}")
```

```python
total_marks = np.sum(marks, axis=1)

for name, total in zip(students, total_marks):
    print(f"{name:<10} Total = {total}")

percentage = (total_marks / 500) * 100

print("Percentage of Each Student\n")

for name, percent in zip(students, percentage):
    print(f"{name:<10} {percent:.2f}%")
```

```python
student_labels = students
student_values = percentage

print("Student Labels")
print(student_labels)

print("\nStudent Percentages")
print(student_values)
```

```python
subject_labels = subjects
subject_values = np.mean(marks, axis=0)

print("Subject Labels")
print(subject_labels)

print("\nAverage Marks in Each Subject")
print(subject_values)
```

```python
import matplotlib.pyplot as plt

plt.bar(student_labels, student_values)
plt.xlabel("Students")
plt.ylabel("Percentage")
plt.title("Student Performance")
plt.show()
```

```python
plt.bar(subject_labels, subject_values)
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.title("Subject-wise Average")
plt.show()
```
