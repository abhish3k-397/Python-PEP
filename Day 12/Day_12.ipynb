| Method                          | Description                                                             |
| ------------------------------- | ----------------------------------------------------------------------- |
| `append(x)`                     | Adds `x` to the end of the list.                                        |
| `extend(iterable)`              | Adds all elements from another iterable to the end.                     |
| `insert(i, x)`                  | Inserts `x` at index `i`.                                               |
| `remove(x)`                     | Removes the first occurrence of `x`.                                    |
| `pop([i])`                      | Removes and returns the element at index `i` (last element if omitted). |
| `clear()`                       | Removes all elements from the list.                                     |
| `index(x[, start[, end]])`      | Returns the index of the first occurrence of `x`.                       |
| `count(x)`                      | Returns the number of times `x` appears.                                |
| `sort(key=None, reverse=False)` | Sorts the list in place.                                                |
| `reverse()`                     | Reverses the list in place.                                             |
| `copy()`                        | Returns a shallow copy of the list.                                     |

Mini Project library Management System

```python
import copy
lib  = list(map(str,input("Enter the Library catalogue: ").split()))
backup = copy.deepcopy(lib)
op = input("What do you want to do: \n 1. Add Book (add) \n 2. Remove Book (rm) \n 3. Search (find) \n 4. Issue (issue) \n 5. Return Book (return) \n 6. create backup (backup) \n 7. Sort book (sort) \n 8. Display the library (Display): (exit)")
while (op != "exit"):
  if op == "add" or op == 1:
    book = input("Enter the book to add: ")
    lib.append(book)
    print(*lib)
  elif op == "rm" or op == 2 :
    book = input("Enter the book to remove: ")
    lib.remove(book)
    print(*lib)
  elif op == "find" or op == 3:
    book = input("Enter the book to find: ")
    if book in lib:
      print(f"Book found at index {lib.index(book)}")
    else:
      print("Book not found")
  elif op == "issue" or op == 4:
    book = input("Enter the book to issue: ")
    if book in lib:
      lib.remove(book)
      print(*lib)
  elif op == "return" or op == 5:
    book = input("Enter the book to return: ")
    lib.append(book)
    print(*lib)
  elif op == "backup" or op == 6:
    backup = copy.deepcopy(lib)
    print(*backup)
  elif op == "sort" or op == 7:
    how = input("Ascending or Descending: ")
    if how == "Ascending":
      lib.sort()
    elif how == "Descending" or op == 8:
      lib.sort(reverse = True)
    print(*lib)
  elif op == "Display" or op == 9:
    print(*lib)
  else:
    print("Invalid input")
  op = input("What do you want to do: \n 1. Add Book (add) \n 2. Remove Book (rm) \n 3. Search (find) \n 4. Issue (issue) \n 5. Return Book (return) \n 6. create backup (backup) \n 7. Sort book (sort) \n 8. Display the library (Display): (exit)")
```
