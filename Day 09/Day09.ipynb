Factorial using recursion

```python
def fact(n):
  if n == 0:
    return 1
  else:
    return n * fact(n-1)
n = int(input("Enter number to get factorial"))
print(fact(n))
```

sum of n

```python
def sum(n):
  if n == 0:
    return 0
  else:
    return n + sum(n-1)
n = (int(input("Enter number to get sum")))
print(sum(n))
```

```python
def find_max(lst):
    maximum = float('-inf')
    for item in lst:
        if isinstance(item, list):
            maximum = max(maximum, find_max(item))
        else:
            maximum = max(maximum, item)
    return maximum

nested = eval(input())
print(find_max(nested))
```

```python
def add(m,n):
  return m+n
def sub(m,n):
  return m-n
def mul(m,n):
  return m*n
def div(m,n):
  return m/n

def calculator(m,n,op):
  if op == '+':
    return add(m,n)
  elif op == '-':
    return sub(m,n)
  elif op == '*':
    return mul(m,n)
  elif op == '/':
    return div(m,n)
  else:
    return "invalid"
m = int(input())
n = int(input())
op = input()
print(calculator(m,n,op))
```

```python
words = [
    "zero", "one", "two", "three", "four",
    "five", "six", "seven", "eight", "nine"
]

def ntw(n):
    if not n:
        return ""

    return words[int(n[0])] + " " + ntw(n[1:])

n = input()
print(ntw(n))
```

```python
def dec_to_bin(n):
    if n == 0:
        return "0"
    if n == 1:
        return "1"

    return dec_to_bin(n // 2) + str(n % 2)

n = int(input())
print(dec_to_bin(n))
```
